```python
import pandas as pd

url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

dados = pd.read_csv(url_dados, compression='zip')
dados
```

```python
mapa = {'droga': 'composto'}
dados.rename(columns=mapa, inplace=True)
dados
```

```python
dados['composto'].value_counts()
```

```python
#pega os 5 primeiros compostos
dados['composto'].value_counts()[0:5]
```

```python
cod_compostos = dados['composto'].value_counts().index[0:5]
cod_compostos
```

```python
# dentro da coluna composto, tudo q tiver contido dentro da variavel cod_composto, vai retornar o dataframe com apenas essas informacoe
dados.query('composto in @cod_compostos')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

dados_compostos = dados.query('composto in @cod_compostos')
# salva as configuracoes so seaborn p deixar o grafico mais bonito
sns.set()
# aumenta o tamanho do grafico
plt.figure(figsize=(10, 8))
ax = sns.countplot(x='composto', data=dados_compostos)
ax.set_title('Top 5 compostos')
plt.show()
```
