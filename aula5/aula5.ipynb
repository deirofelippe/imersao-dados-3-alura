```python
import pandas as pd

url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

dados = pd.read_csv(url_dados, compression='zip')

url_dados_resultados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_resultados.csv?raw=true'

dados_resultados = pd.read_csv(url_dados_resultados)

dados_resultados
```

```python
dados_resultados['n_moa'] = dados_resultados.drop('id', axis=1).sum(axis=1)
```

```python
dados_resultados['ativo_moa'] = (dados_resultados['n_moa'] != 0)
dados_resultados
```

```python
dados_merge = pd.merge(dados, dados_resultados[['id','n_moa','ativo_moa']], on='id')
dados_merge
```

```python
dados_merge.query('tratamento == "com_controle"')['ativo_moa'].value_counts()
```

```python
dados_merge.query('tratamento == "com_droga"')['ativo_moa'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

## Regressão Logística

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

x = dados_merge.select_dtypes('float64')
y = dados_merge['ativo_moa']
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.2, stratify=y, random_state=376)

modelo_rlogistica = LogisticRegression(max_iter=1000)
modelo_rlogistica.fit(x_treino, y_treino)
modelo_rlogistica.score(x_teste, y_teste)
```

```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score

modelo_dummy = DummyClassifier(strategy='most_frequent')
modelo_dummy.fit(x_treino, y_treino)
previsao_dummy = modelo_dummy.predict(x_teste)
accuracy_score(y_teste, previsao_dummy)
```

## Árvore de decisão
Arvore de decisao tem o problema de overfit, ele fica tao bom em advinhar os dados de treino que perde a capacidade de generalizar. Uma das maneiras de evitar o overfit é criar varias arvores de decisoes, que esse modelo é chamado de **random forest**

```python
from sklearn.tree import DecisionTreeClassifier

x = dados_merge.select_dtypes('float64')
y = dados_merge['ativo_moa']
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.2, stratify=y, random_state=376)

modelo_arvore = DecisionTreeClassifier(max_depth=3)
modelo_arvore.fit(x_treino, y_treino)
modelo_arvore.score(x_teste, y_teste)
```

```python
from sklearn import tree

fig, ax = plt.subplots(figsize=(15,10), facecolor='k')
tree.plot_tree(modelo_arvore,
                ax=ax,
                fontsize=10,
                rounded=False,
                filled=True,
                feature_names=x_treino.columns,
                class_names=['Não Ativado', 'Ativado'])
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier

x = dados_merge.select_dtypes('float64')
y = dados_merge['ativo_moa']
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.2, stratify=y, random_state=376)

teste=[]
treino=[]
for i in range(1,15):
    modelo_arvore = DecisionTreeClassifier(max_depth=i)
    modelo_arvore.fit(x_treino, y_treino)
    teste.append(modelo_arvore.score(x_teste, y_teste))
    treino.append(modelo_arvore.score(x_treino, y_treino))
```

```python
teste
```

```python
treino
```

```python
sns.lineplot(x=range(1,15), y=teste, label='teste')
sns.lineplot(x=range(1,15), y=treino, label='treino')
```

```python
from sklearn.ensemble import RandomForestClassifier

x = dados_merge.drop(['id','n_moa','ativo_moa','droga'], axis=1)
x = pd.get_dummies(x, columns=['tratamento','dose','tempo'])
y = dados_merge['ativo_moa']
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.2, stratify=y, random_state=376)

modelo_dandomforest = RandomForestClassifier()
modelo_dandomforest.fit(x_treino, y_treino)
modelo_dandomforest.score(x_teste, y_teste)
```
